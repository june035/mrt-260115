# 파이썬 함수 (function)
- 함수 : 어떤 입력값을 받아서 내부에서 연산처리 후 해당 결과값을 외부로 출력할 수 있도록 기능하는 것
          (프로그래밍 언어의 꽃)
- 함수에 값을 입력 & 출력 : input & output
- 함수 생성(=선언) -> 함수 존재 -> 값(=인자값=인수=매개변수) 입력 *값이 들어갈 수도, 없을 수도 있다
- 함수 사용(=호출)
- 함수의 선언부 & 호출부의 값은 갯수는 기본적으로 같아야한다.
    - 어느 특정 한 쪽이 상대적으로 갯수가 많거나 적으면 정상적으로 실행되지 않는다.
- return 문 : 함수의 내부 연산 처리 후 특정 값을 함수 바깥으로 값을 보낼 수 있는 기능

*func1()은 임의로 지정한 함수의 이름. 함수가 들어가는 위치
*: 아래를 실행한다

def func1() :
    실제 함수의 실행연산이 진행될 공간 -> 반복문, 조건문 사용

- 함수는 반복적인 실행코드를 자동화하고 싶을 때 사용한다.

# 파이썬 자료 구조
- 문자열
- 정수
- 실수
- 논리형
> 단일값을 저장하는 자료 형태. 변수 1 : 값 1

- 리스트 = [1, 2, 3] -> 원하는 값을 추가, 삭제 가능.
- 튜플 = (1, 2) 소괄호가 나오면 우선 튜플을 의심해야 함.
    - 리스트와 유사하지만 차이점: 최초에 선언된 형태의 자료 구조 변경 불가. 데이터에 추가, 삭제 불가.
    - 튜플은 거의 사용하지 않음. b/c 데이터 수집은 변동성을 담보로 하는데 튜플은 100% 고정값이어야 하므로. (100% 고정값=주민번호, 성별 같은)
> 리스트와 튜플은 iter 속성을 가져 반복문 가능
> 단, 리스트와 튜플은 그 안에 있는 담겨있는 값들이 구체적으로 어떤 값을 의미하는지 알 수 없다.

- 딕셔너리: {"apple": "사과"} -> apple = "key" / 사과 = "value" -> key+value = property = 속성 = {"apple": "사과"}
    - 한 쌍의 값을 찾아오려고 할 때
    - 반대로 단일값을 찾아올 때는 리스트
    - 딕셔너리 자주 사용하는 예시 : 인터파크 공연 페이지 데이터 수집 크롤링
                            리스트를 사용하면 공연명과 가수를 매칭할 수 없음
                            {"싸이": "워터밤", "성시경": "사랑합니다"} VS ["워터밤", "사랑합니다"]
    - 리스트 사용 예시 : 댓글 크롤링
                          ["매장이 넓어요", "화장실이 깨끗해요"]
- 세트(집합)
    - 교집합, 합집합, 차집합, 공집합 등 그룹화
    - 반복되어지는 값을 허용하지 않음
    - 데이터를 크롤링한 값을 찾아올 때, 반복적으로 입력된 값이 들어온다면 1회 처리
        - 예시: 인스타그램 > 데이터 크롤링 > 네일아트 업체에서 똑같은 컨텐츠 반복 게시하면 모두 들어올텐데, 집합을 쓰면 1개만 들어옴 

> 다중값을 저장할 수 있는 자료 형태. 변수 1 : 값 3

> 대괄호 [] = 리스트, 중괄호 {} = 딕셔너리/집합, 소괄호 ()= 튜플
> 두 번째 값은 사용 빈도 적음

1. 리스트 : 변수명 = [값]
> 변수명 = list() 도 가능

2. 튜플 : 변수명 = (값)
> 변수명 = tuple() 도 가능

3. 딕셔너리 : 변수명 = {"key": "value"}
> 변수명 = dict() 도 가능

4. 세트 : 변수명 = {}
> 변수명 = set({})
> 변수명 = set()

```python
def test_func1(a, b) :
    quot = a // b # // = quot = 나누기의 몫
    rem = a % b # % = rem = 나누기의 나머지
    return (quot, rem)
```

```python
result = test_func1(10, 3)
# result

# 튜플은 구조 분해 할당 기능을 가졌다. 
# (f, s) = test_func1(10,3) 
(f, s) = result

print(f, s)
```

```python
data_dict = {"한국": "KR", "일본": "JP", "중국": "CN"}
print(type(data_dict))
```

```python
# 리스트, 튜플은 내부 아이템에 고유한 인덱스(0부터 시작)를 가졌다. 따라서 인덱스를 찾으면 그 값을 가져올 수 있었음

# print(dir(data_dict)) 'iter'의 속성을 가졌음
# data_dict[0] 그렇지만 딕셔너리는 인덱스값을 찾아올 수 없음. 한 쌍(프로퍼티)의 속성을 가졌기 때문

print(data_dict["한국"], data_dict["일본"]) # 'key' 값을 입력해야한다
```

```python
# 딕셔너리 자료 구조는 iter 속성을 가졌다. 

for item in data_dict :
    # print(item) 딕셔너리는 key 를 중심으로 전개
    # print(data_dict[item]) 딕셔너리는 [key] 로 value를 찾아올 수 있다. 
    # print(item, data_dict[item])
    print(f"{item} = {data_dict[item]}")
```

```python
# 그렇다면, key 와 value 를 딕셔너리에서 찾아오려면 매번 반복문을 사용해야 하나?

# dir(data_dict) #해당 자료구조가 사용할 수 있는 속성 혹은 메서드 함수의 목록 dir(). keys, values, items(keys+values)를 가짐

data_dict.keys()
# type(data_dict.keys())
```

```python
for values in data_dict.values() :
    print(values)
    
for key in data_dict.keys() :
    print(data_dict[key])
```

```python
# set 자료구조는 중복값을 허용하지 않는다

data_list = ["apple", "dell", "samsung", "lg", "apple", "lg", "apple", "samsung"]
```

```python
data_list
```

```python
set_result = set(data_list) #명시적 형 변환 VS 자동 형 변환
print(set_result)
```
