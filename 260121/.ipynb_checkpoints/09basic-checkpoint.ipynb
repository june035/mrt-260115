> 예시: "냉장고에 코끼리를 넣는 방법"을 작성한다면
> 코끼리 라는 대상의 정보를 메뉴얼(Class)로 만든다 -> 해당 메뉴얼을 기반으로 코끼리1(객체)을 만든다 -> 냉장고라는 대상의 정보를 메뉴얼로 만든다 -> 해당 메뉴얼을 기반으로 냉장고1(객체)를 만든다 -> 코끼리1 객체가 가지고 있는 기능을 사용해서 냉장고1 객체의 문을 열게 한다 -> 코끼리1 객체가 가지고 있는 기능을 사용해서 냉장고1 객체 안으로 들어가게 한다 -> 코끼리1 객체가 가지고 있는 기능을 사용해서 냉장고1 객체의 문을 닫게 한다

```python
# 우리가 직접 만든 독립적인 함수
def print_string(data) :
    print(data)
```

```python
# 클래스를 통해서 만들어진 메서드 함수. 
string = "David Dave"
print(string.split())
```

```python
class Student : #클래스 선언
    def __init__(self, name, grade) : #생성자 함수
        self.name = name
        self.grade = grade
        
    def study(self) : #메서드 함수
        print(self.name, "공부합니다.")
        
    def take_exam(self, subject) : #메서드 함수
        print(self.name, subject, "공부합니다.")
```

```python
student1 = Student("David", 3)
student2 = Student("Jane", 2)

print(student1.name, student1.grade)
print(student2.name, student2.grade)
```

```python
student1.study()
```

```python
student1.take_exam("수학")
```
