```python
price = "100 달러"
exchange_rate = 1480

# price 라는 변수에 있는 100달러의 값을 환율을 적용해서 한화로 변환해서 값을 출력해주세요.

dollar = 100
won = dollar * 1480
print(f"{won}원")
```

```python
price = "100 달러"
exchange_rate = 1480

#문자열 : 인덱스를 가졌으므로 ***슬라이싱*** 기능을 사용
print(int(price[0:3]))

result = int(price[0:3]) * exchange_rate
print(result)
print(f"{result:,}원") # 값:, 1000의 단위마다 쉼표를 찍어라
```

```python
# 반복문을 활용해서 구구단 2~9단까지를 아래에 출력해주세요.

# 1. 2단부터 9단까지 반복한다 for index_1 in range(2, 10) 2부터 시작해서 9까지

for index_1 in range(2, 10) :
    print(f"==== 구구단 {index_1}단====")
    for index_2 in range(1, 10) :
        print(f"{index_1} x {index_2} = {index_1 * index_2}")
    print("\n") #\n 행을 바꾸겠다. print("\n") escape sequnce, 문자처럼 생겼지만, 문자 이상의 고유한 문법을 가지고 있는 문자
```
