```python
# 상품명 / 카테고리명 가져오기

import requests
from bs4 import BeautifulSoup

url = "https://crawlingtest-0122.netlify.app/"
headers = {
    "User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/143.0.0.0 Safari/537.36"
}

res = requests.get(url, headers=headers)
res.encoding = "utf-8"

soup = BeautifulSoup(res.text, "html.parser")

result = soup.select(".product-card")

for r in result :
    name = r.select_one(".product-name").text.strip()
    category = r.select_one(".category").text.strip()
    print(f"{name} / {category}")
```

```python
# 상품당 가격 출력
import requests
from bs4 import BeautifulSoup

url = "https://crawlingtest-0122.netlify.app/"
headers = {
    "User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/143.0.0.0 Safari/537.36"
}

res = requests.get(url, headers=headers)
res.encoding = "utf-8"

soup = BeautifulSoup(res.text, "html.parser")

for card in soup.select(".product-card") :
    name = card.select_one(".product-name").text.strip()
    price = int(card.select_one(".price")["data-price"])
    print(f"{name} : {price:,}")
```

```python
# 5만원 이상의 상품의 이름만 크롤링
import requests
from bs4 import BeautifulSoup

url = "https://crawlingtest-0122.netlify.app/"
headers = {
    "User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/143.0.0.0 Safari/537.36"
}

res = requests.get(url, headers=headers)
res.encoding = "utf-8"

soup = BeautifulSoup(res.text, "html.parser")

for card in soup.select(".product-card") :
    price = int(card.select_one(".price")["data-price"])
    if price >= 50000 :
        name = card.select_one(".product-name").text.strip()
        print(f"{name} / {price:,}")
```

```python
# Q. 특정 사이트에 접속해서 데이터를 수집 + 재활용 (수집된 데이터를 기반으로 프로토타입 생성)
# 수집한 상품별 상품명, 각각의 카테고리, 가격을 하나의 딕셔너리로 만들고, 해당 딕셔너리가 하나의 리스트 안에 저장될 수 있도록 해주세요.
"""
[
{
"상품명": "..."
"카테고리": "..."
"price": "..."
},
{},
{}
]
"""

import requests
from bs4 import BeautifulSoup

url = "https://crawlingtest-0122.netlify.app/"
headers = {
    "User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/143.0.0.0 Safari/537.36"
}

res = requests.get(url, headers=headers)
res.encoding = "utf-8"

soup = BeautifulSoup(res.text, "html.parser")

items = []

for card in soup.select(".product-card") :
    items.append({
        "name" : card.select_one(".product-name").text.strip(),
        "category" : card.select_one(".category").text.strip(),
        "price" : int(card.select_one(".price")["data-price"])
    })
print(items)
```

```python
import requests
from bs4 import BeautifulSoup

res = requests.get("https://crawlingtest.netlify.app/#")
soup = BeautifulSoup(res.text, "html.parser")

section = soup.find("ul", id="dev_course_list")
titles = section.find_all("li", "course")

for index, title in enumerate(titles) :
    print(index + 1, "댓글 수 : ",title.get_text(" ", strip=True).split(" - ")[-1].replace("\n", "").replace(" ", "")[-2:-1])
```
