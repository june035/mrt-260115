```python
from bs4 import BeautifulSoup 

html = """ 
<p class="desc">
    여름용
    <strong>오프화이트</strong>
    반팔 티셔츠
</p>
"""

soup = BeautifulSoup(html, "html.parser")
# desc = soup.select_one(".desc").text.strip() 앞/뒤 영역의 공백에만 영향
desc = soup.select_one(".desc").get_text(" ", strip=True) #내부 영역의 공백까지도 영향
print(desc)
```

```python
from bs4 import BeautifulSoup 

html = """ 
<div class="product-card">
    <h2 class="product-name">나이키 러닝화</h2>
    <p class="category">신발</p>
    <p class="price" data-price="129000">129,000원</p>
</div>
"""

soup = BeautifulSoup(html, "html.parser")

card = soup.find("div", class_="product-card")
name = card.find("h2", class_="product-name").text.strip()
price = card.find("p", class_="price")["data-price"] # find & select_one : 문서 위에서부터 스캔, 매칭값 나타나면 스캔 X
print(price)
```
