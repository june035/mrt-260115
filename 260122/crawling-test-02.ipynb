### 파이썬 requests 를 활용한 크롤링 팁
패턴학습
1. 단일값을 찾는 경우는 없음. 복수의 값을 가지고 있는 부모 찾기 > 부모 찾고 자식 요소 찾기
- 자주 등장하는 패턴: ul > li
- ul = unorder list = 순번이 정렬되지 않은 리스트
- ul은 무조건 복수의 li를 가진다

2. 리스트 안에서 인덱스와 슬라이싱 활용

```python
import requests
from bs4 import BeautifulSoup

res = requests.get("https://crawlingtest.netlify.app/#")
soup = BeautifulSoup(res.text, "html.parser")

# print(soup)

# section = soup.select(".course.maintitle")
# print(section)

section = soup.find("ul", id="dev_course_list")
titles = section.find_all("li", "course")

# for title in titles :
#   print(title.get_text(" ", strip=True))

for index, title in enumerate(titles) :
    print(index + 1, "댓글 수 : ", title.get_text(" ", strip=True).split(" - ")[-1].replace("\n", " ").replace(" ", "")[-2:-1])
```

```python
import requests
from bs4 import BeautifulSoup

res = requests.get("https://crawlingtest.netlify.app/#")
soup = BeautifulSoup(res.text, "html.parser")

items = soup.select("ul > li")
for item in items :
    print(item.get_text(" ", strip=True).replace("\n", "").replace(" ", ""))
```

1. 해당 사이트에 접속해서 https://quotes.toscrape.com/
2. 명언, 사람, 태그를 크롤링
3. 위 3개 요소를 크롤링한 이후 해당 값을 딕셔너리 자료 형태로 묶어주세요.
4. 그리고 해당 딕셔너리를 감싸는 리스트를 만들어서 최종적으로 출력해주세요.

```python
import requests
from bs4 import BeautifulSoup

url = "https://quotes.toscrape.com/"
headers = {
    "User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/143.0.0.0 Safari/537.36"
}

res = requests.get(url, headers=headers)
res.encoding = "uft-8"

soup = BeautifulSoup(res.text, "html.parser")

information = []

for i in soup.select(".col-md-8 > .quote") :
    information.append({
        "Quote" : i.select_one(".text").text.strip(),
        "Author" : i.select_one("small.author").text.strip(),
        "Tags" : i.select_one(".tags").text.strip().split("Tags:")[-1].replace("\n", "").replace(" ", "")
    })
print(information)
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://quotes.toscrape.com/"
headers = {
    "User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/143.0.0.0 Safari/537.36"
}

res = requests.get(url, headers=headers)
res.encoding = "uft-8"

soup = BeautifulSoup(res.text, "html.parser")

information = []

for i in soup.select(".quote") :
    information.append({
        "Quote" : i.select_one(".text").text.strip(),
        "Author" : i.select_one(".author").text.strip(),
        "Tags" : i.select_one(".tags").text.strip().replace("\n", " ").split(":")[1]
    })
print(information)
```
