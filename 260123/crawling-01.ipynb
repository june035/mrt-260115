```python
# HTTP(Hyper Text Transfer Protocol) 상태코드
# 참고 : https://namu.wiki/w/HTTP/%EC%9D%91%EB%8B%B5%20%EC%BD%94%EB%93%9C

# 서버 측에서 응답 메세지를 의도적으로 차단하거나 오류로 발생시킬 수 있다.
# 이럴 때는 예외 조항 처리

# 크롤링을 컴퓨터가 정상적으로 진행하지 못할 때 원인 찾기

import requests
from bs4 import BeautifulSoup

url = "https://davelee-fun.github.io"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36"
}

response = requests.get(url, headers=headers)

# '예외조항처리' 서버와의 통신 확인하기. 
if response.status_code != 200 :
    print("페이지 없음")
else :
    soup = BeautifulSoup(response.content, "html.parser") 
    data = soup.select(".card-text")
    for item in data :
        print(item.get_text(" ", strip=True))
```

```python
# 복수의 페이지 크롤링 하는 방법
# 1) 각 페이지별 주소창의 주소값을 확인한다
# 2) 페이지별 주소값의 차이점을 확인한다. 즉, 패턴 확인
# 3) 확인한 패턴에 부합하는 파이썬 코드를 작성해서 페이지별 크롤링을 진행한다.

# https://davelee-fun.github.io/index.html
# https://davelee-fun.github.io/page2/
# https://davelee-fun.github.io/page3/
# https : 프로토콜(통신규약)
# davelee-fun.github.io : 도메인 주소
# /page2/ : 파라미터 => 경로 정의
# /index.html : 기본 페이지 의미. =없어도 된다는 뜻

```

```python
# 8페이지까지 
import requests
from bs4 import BeautifulSoup

url = "https://davelee-fun.github.io"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36"
}

response = requests.get(url, headers=headers)

if response.status_code != 200 :
    print("페이지 없음")
else :
    # 8회의 반복. 0부터 7까지 총 8회 반복. 
    for page_num in range(8) :
        if page_num == 0 : # 1페이지
            response = requests.get("https://davelee-fun.github.io", headers=headers)
        else : #2~8페이지
            response = requests.get(f"https://davelee-fun.github.io/page{str(page_num + 1)}", headers=headers)
            # response = requests.get("https://davelee-fun.github.io/page" + str(page_num + 1), headers=headers)

        soup = BeautifulSoup(response.content, "html.parser") 
        data = soup.select(".card-text")
        for item in data :
            print(item.get_text(" ", strip=True))
```

```python
import openpyxl

def write_excel_template(filename, sheetname, listdata) :
    excel_file = openpyxl.Workbook()
    excel_sheet = excel_file.active #active : 생성되어진 엑셀파일에 기본 시트를 사용하겠다
    excel_sheet.column_dimensions["A"].width = 100 #100개 정도의 단어가 들어갈 A열을 만들겠다

    if sheetname != "" :
        excel_sheet.title = sheetname

    for item in listdata :
        excel_sheet.append(item) #엑셀시트 자체가 리스트이므로 append 사용
    
    excel_file.save(filename)
    excel_file.close()
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://davelee-fun.github.io"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36"
}

response = requests.get(url, headers=headers)
product_lists = list() #product_lists = [] 과 동일

if response.status_code != 200 :
    print("페이지 없음")
else :

    # 8회의 반복. 0부터 7까지 총 8회 반복. 
    for page_num in range(8) :
        if page_num == 0 : # 1페이지
            response = requests.get("https://davelee-fun.github.io", headers=headers)
        else : #2~8페이지
            response = requests.get(f"https://davelee-fun.github.io/page{str(page_num + 1)}", headers=headers)
            # response = requests.get("https://davelee-fun.github.io/page" + str(page_num + 1), headers=headers)

        soup = BeautifulSoup(response.content, "html.parser") 
        data = soup.select(".card")
        for item in data :
            product_name = item.select_one(".card-text")
            product_date = item.select_one(".post-date")
            product_info = [product_name.get_text(" ", strip=True), product_date.get_text(" ", strip=True)]
            product_lists.append(product_info)
```

```python
write_excel_template("tmp.xlsx", "상품정보", product_lists)
```
