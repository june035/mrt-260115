### API 를 활용한 데이터 수집 방법

- = Application Programming interface
- 어플리케이션을 프로그래밍하는데 있어서 처음부터 개발하지 않고 기존 서비스의 측면에 접속해서 데이터를 가져온 후, 그 데이터를 기반으로 프로그래밍을 하겠다
- 예시1) 네이버 사이트 안에 오늘의 날씨 안내 정보 > 네이버는 기상청이 아님에도 어떻게 정보 제공?

```python
# 데이터 1000개 추출, 엑셀파일 저장

import requests
import openpyxl

client_id = "IOMSg4qqzvK2W40F4hc5"
client_secret = "bvwDXgPDwu"

excel_file = openpyxl.Workbook() #엑셀 파일 만듦
excel_sheet = excel_file.active # 첫번째 시트 활성화
excel_sheet.column_dimensions["B"].width = 100 #B열 넓이 조정
excel_sheet.column_dimensions["C"].width = 100
excel_sheet.append(["No", "제목", "링크"]) #첫번째 행에 들어갈 내용

# A열: 1~1000, B열: 제목, C열: 상품 링크

start = 1
num = 0

for index in range(10) :
    start_number = start + (index * 100)

    naver_open_api = f"https://openapi.naver.com/v1/search/shop.json?query=독도토너&display=100&start={str(start_number)}"
    header_params = {
        "X-Naver-Client-Id": client_id,
        "X-Naver-Client-Secret": client_secret
    }
    
    res = requests.get(naver_open_api, headers=header_params)
    
    if res.status_code == 200 :
        data = res.json()
        for item in data["items"] :
            num += 1 #1~1000까지 넘버링. enumerate 는 1~100을 하고, 다시 1~100. 101~1000으로 안 감.
            excel_sheet.append([num, item["title"], item["link"]])
    else :
        print("Error Code :", res.status_code)

excel_file.save("shopping_with_독도토너.xlsx")
excel_file.close()
print("shopping_with_독도토너.xlsx 파일 저장 완료 Good job!")
```
