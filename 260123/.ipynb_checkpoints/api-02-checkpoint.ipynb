```python
import requests

url = "https://apis.data.go.kr/B551011/GoCamping/basedList"
request_params = {
    "numOfRows":"10",
    "pageNo":"1",
    "MobileOS":"ETC",
    "MobileApp":"camping",
    "serviceKey":"42a5635aa5c75221fc6de35f3eb03327ef3c1c8502f63839d7084ddb87d1a3b9",
    "_type":"xml"
    
}

response = requests.get(url, params=request_params)
print(response)
```

```python
# 위 코드가 안 먹힐 때. 정상적인데 404 뜰 때.

import subprocess
import xml.etree.ElementTree as ET

url = "https://apis.data.go.kr/B551011/GoCamping/basedList?numOfRows=50&pageNo=1&MobileOS=ETC&MobileApp=camping&serviceKey=42a5635aa5c75221fc6de35f3eb03327ef3c1c8502f63839d7084ddb87d1a3b9&_type=xml"

result = subprocess.run(["curl", "-k", url], capture_output=True, text=True, encoding="utf-8")
root = ET.fromstring(result.stdout)

print(root.findall(".//item")) #. 최상위에서 // 그 아래에 있는 item 찾아와
```
