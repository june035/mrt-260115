API를 활용한 데이터 수집 방법
- Application Programming interface
- 어플리케이션을 프로그래밍하는데 있어서 처음부터 개발하지 않고 기존 서비스의 측면에 접속해서 데이터를 가져온 후, 그 데이터를 기반으로 프로그래밍을 하겠다
- 예시1) 네이버 사이트 안에 오늘의 날씨 안내 정보 > 네이버는 기상청이 아님에도 어떻게 정보 제공?
- 예시2) 카카오맵이 버스, 지하철 도착 시간과 빈 좌석 정보 제공 > 어떻게?
- ==> 정부 기관에서 제공하는 대중교통 정보를 수집하고, 카카오가 데이터를 가져가서 서비스하는 것.
- ==> 결국 어떤 서비스를 만든다는 이야기 . CRUD
- Create / Read / Update / Delete

JSON
- Javascript Object Notation = 자바스크립트 객체 표기법
- 웹사이트는 html, css, js로 구성되어 있는데, 엄밀하게 js로 만들어짐.
- oop : Object-Oriented Programming. 객체 지향 언어로 만들어짐. 객체 지향(중심) 프로그래밍(컴퓨터에게 시키는 일). 객체(물건)을 중심으로 생각하면서 프로그램을 만든다. 프로그래밍 사고방식. 프론트엔드 진영의 프로그래밍 대표 언어
- Java : 프로그래밍 언어(컴퓨터한테 시키는 말). 백엔드 진영의 프로그래밍 대표 언어. 동일하게 객체지향언어
- 객체지향 언어를 최대한 보존하면서 프론트&백엔드가 소통할 수 있게 만든, 공용 언어(형식, 데이터 모양) json.
- {"name": "David"} = 파이썬에서 딕셔너리 = JSON
