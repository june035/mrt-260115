조건문 : 값을 출력하기 위해서 A | B 경우에 따라서 다른 값을 출력해야 할 때

1. if 조건문 : A라면 ~ 해라 / B라면 ~해라 (단일 조건)
2. if else문 : A라면 ~ 해라, 만약 A가 아니라면 - 해라 (양자택일)
3. if elif else문: A라면 ~, B라면, C라면 (다중택일)

- else문: 무조건 if, elif보다 먼저 나올 수 없다. 비필수조건
- 조건문의 조건식에 비교연산자, 논리연산자가 주로 사용됨!

- 중첩 if 조건문

```python
"""
if 조건식 :
    실행문

if 조건1 and 조건2 :
    실행문

if not 조건 : -> 조건식의 값이 부정(False)일 경우에 한해서 실행문 실행 시켜야하는 경우
    실행문

if 조건식1 :
    실행문1
else :
    if 조건2 :
        실행문2
    else :
        실행문3
"""
```

```python
num1 = int(input("첫 번째 숫자"))
num2 = int(input("두 번째 숫자"))
num3 = int(input("세 번째 숫자"))

if num1 <= num2 and num1 <= num3 :
    print(num1)
elif num2 <= num1 and num2 <= num3 :
    print(num2)
else :
    print(num3)
```

```python
numbers = []

for i in range(3) :
    number = int(input(f"{i+1}번째 숫자 : "))
    numbers.append(number)

num1, num2, num3 = numbers

if num1 <= num2 and num1 <= num3 :
    print(num1)
elif num2 <= num1 and num2 <= num3 :
    print(num2)
elif num3 <= num1 and num3 <= num2 :
    print(num3)
```

```python
numbers = []

for i in range(3) :
    number = int(input(f"{i+1}번째 숫자 : "))
    numbers.append(number)

num1, num2, num3 = numbers

print(min(numbers))
print(max(numbers))
```

```python
numbers = [int(input(f"{i + 1}번째 숫자 : ")) for i in range(3)]

print(numbers)
print(min(numbers))
print(max(numbers))
print(sorted(numbers)[0])
```

```python
id_num = input("주민번호는?")

gender = id_num.split("-")[1][0]

if gender == "1" or gender == "3" :
    print("당신은 남성입니다")
elif gender == "2" or gender == "4" :
    print("당신은 여성입니다")
else :
    print("잘못 입력하셨습니다.")
```
