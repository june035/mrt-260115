-출력: print() 내장함수=built in function
-입력: input("") placehloder
    - 해당 함수를 통해서 수집된 값을 무조건 "문자열"이 된다

-자료형: 변수 안에 담을 수 있는 값의 형태. 문자열(String), 문자, 정수(Inteaser, Int), 실수, 논리형(Boolean)
    - 형 변화: 자동형 변환 - input() / 명시적형 변환 - 아래 함수들을 통해 수동적으로 형태 변환
    - int() : 정수형 변환
    - float() : 실수형 변환
    - str() : 문자열 변환
    - bool() : 논리형 변환
    - list() : 리스트 자료 구조 변환
    - dict() : 딕셔너리 자료 구조 변환

```python
print("Hello Wolrd")
print('Hello Wolrd')
print(1+2)
print((1+2)*3)

num1 = 4 
num2 = 2

print(num1 + num2)
print(num1 - num2)
print(num1 * num2)
print(num1 / num2)
print(num1 % num2)
print(num1 ** num2)

num1 = 8
print(num1 + num2)
```

```python
num1 = 8
num2 = 2

print(num1 > num2)
print(num1 < num2)
print(num1 >= num2)
print(num1 <= num2)
print(num1 == num2)
print(num1 != num2)
```

```python
radius = 10
pi = 3.1415

print("지름 :", radius*2)
print("원의 둘레 :", 2*pi*radius)
print("원의 넓이 :", pi*(radius**2))
print(f"원의 넓이 : {pi*radius*radius}")
```
