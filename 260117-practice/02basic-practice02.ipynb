```python
str = "I have a {}, I have an {}"

print(str.format("pen", "apple"))
print("I have a {1}, I have an {0}".format("apple", "pen"))

interest = 0.087
print(format(interest, ".2f"))
print(f"{interest:.2f}")
```

```python
import math

interest = 0.891
result01 = math.floor(interest)
print(result01)

result02 = math.ceil(interest)
print(result02)
```
