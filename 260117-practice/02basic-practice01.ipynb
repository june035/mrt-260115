- 출력 : print() 내장함수=built in function
- 입력 : input() = placeholder. 입력해야 될 값의 정보
    - 해당 함수를 통해서 수집한 값은 무조건 "문자열"이 된다
- 연산 :
    - 사칙연산: + - * / % ** 유의사항 '+': 문자 간의 연결 연산자로서도 사용 가능
    - 비교연산: < > <= >= == != (우항의 값을 좌항으로 삽입)
- 변수 :
    - variable 변할 수 있는 값을 넣는 그릇
    - 팁: 스네이크 표기법(case_study), 카멜표기법(caseStudy)
- 자료형 : 변수 안에 담을 수 있는 값의 형태. 문자열(String) " ", 문자(Character) ' ', 정수(Int, Inteaser), 실수(Float), 논리형(Boolean)
    - 형 변화:
        - 자동형 변환-input()
        - 명시적형 변환-수동적으로 형태를 변환. 아래 함수들
          - int() : 정수형 변환
          - float() : 실수형 변환
          - str() : 문자열 변환
          - bool() : 논리형 변환
          - list() : 리스트 자료 구조 변환
          - dic() : 딕셔너리 자료 구조 변환
          - tuple() : 튜플 자료 구조 변환
          - set() : 집합 자료 구조 변환
- 문자열 : 독립적인 함수 혹은 속성들이 존재
  - count() : 특정 문자열에서 어떤 단어가 몇 번 사용되었는지를 집계. 대/소문자 구분 필요.
  - len() : lengh. 띄어쓰기도 수량 포함
  - find() : 해당 문자열에서 인자값으로 전달된 문자가 몇 번째 인덱스(0부터 시작)에 위치하는지 찾고자 할 때. 문서에서 가장 처음 찾은 문자의 인덱스 반환
  - replace() : 특정 문자열을 대체하고자 할 때
  - index : 문자열 안에 각각의 문자는 고유한 숫자를 가지고 있다. 문자열 왼쪽부터 숫자를 세며, 0부터 시작.
  - slicing : 문자열 안에서 특정 구간을 잘라내서 가져오고 싶을 때
  - strip() : 특정 문자열 앞뒤에 여백 혹은 여백으로 인지할 수 있는 escape sequence(\n, \t) 문자열을 제거하고자 할 때
  - format() : 특정 문자열을 원하는 위치에 출력하고자 할 때 사용하는 method 함수
- 형식 지정자 : 특정 타입의 값을 출력하고자 할 때, 어떤 형태로 값을 찾아와서 출력할지 결정
    - %s - string
    - %c - character
    - %d - int
    - %f - float

```python
name = "I have a {}, I have an {}"

print(name.format("pen", "apple"))
print("I have a {}, I have an {}".format("pen", "apple"))
print("I have an {1}, I have a {0}".format("pen", "apple"))

interest = 0.087
print(format(interest, ".2f"))
print(f"{interest:.2f}")
```

```python
# 파이썬 내장함수 가운데 무조건 소수점 자리를 올리거나 내리는 독립 기능 없음
# 외부 라이브러리 모듈을 찾아와서 사용해야 함

import math

interest = 0.891
result01 = math.floor(interest)
print(result01)

result02 = math.ceil(interest)
print(result02)
```
