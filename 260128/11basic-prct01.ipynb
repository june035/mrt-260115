```python
### 정규표현식 regular expression

- 어떤 문자열 속에서 내가 원하는 문자를 추출해오고 싶을 때
- 패턴 중심 : 일관성을 가지고 있는 형태

1) [0-9] = \d
2) ^ (캐럿) : 무조건 ~~ 으로 시작한다 예시) ^[0-9] : 숫자로 시작한다
3) [^0-9] : 부정 - 숫자가 아닌 것을 찾아라 \D
4) [\t] : tab 여백 [\n] : new line 개행 [\r] : carriage return : 커서를 맨 앞으로 이동 [\f] : form feed 다음 페이지 [\v] : vertical tab 행간 (세로 여백)
5) [^\t\n\r\f\v] : white space 아닌 것을 찾아와라
6) [A-Za-z0-9], [가-힣], [가-힣A-Za-z0-9]
7) re.sub(패턴, 바꿀 데이터, 원본데이터) 
8) . : 문자 1개를 의미함. D.A -> D () A
9) ? : 물음표를 중심으로 앞에 있는 문자가 0번 또는 1번 표시되는 패턴 D?A -> A 또는 DA
10) * : asterisk 애스터리스크를 중심으로 앞에 있는 문자가 0번 또는 그 이상 [A-Z]* 앞에 글자가 몇 개 올지 모름
11) + : 더하기를 중심으로 앞에 있는 문자가 1번 또는 그 이상
12) {n} : 앞에 있는 문자가 n번 반복되는 패턴을 찾아라 "AD{2}A" = ADDA
13) {m,n} : 앞에 있는 문자가 m번 반복되는 패턴부터 n번 반복되는 패턴까지 범위를 찾아라. m과 n사이에 띄어쓰기 X
14) [ ] : 대괄호 안에 포함된 문자는 허용하겠다
15) search() : 왼쪽부터 패턴에 매칭되는 문자를 찾다가 최초 찾는 순간 정지
16) findall()은 문자열 모두를 끝까지 찾음. 리스트 형식으로 값을 반환
```

```python
import requests
from bs4 import BeautifulSoup
import re

url = "https://crawlingtest.netlify.app/"
res = requests.get(url)
res.raise_for_status

soup = BeautifulSoup(res.text, "html.parser")
datas = soup.select("#dev_course_list .course")

for data in datas :
    text = data.get_text(" ", strip=True)
    clean_text = re.sub(r"\[[0-9]+\]", "", text)
    print(clean_text)
```

```python
import re

text = "문의: help@naver.com, admin@gmail.com, sales@myshop.co.kr"

emails = re.findall(" [A-Za-z0-9.-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}", text)
domains = [e.split("@")[1] for e in emails]
print(domains)
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://davelee-fun.github.io/blog/crawling_stock_example.html"
res = requests.get(url)
soup = BeautifulSoup(res.text, "html.parser")

rows = soup.select("ul.table-stock.sty1 > li.row_sty")
index = 0

for row in rows :
    company = row.select_one("div.st_name > span.name").get_text(" ", strip=True).replace("\n", "").replace(" ", "")
    price = row.select_one("div.st_price > span.price").get_text(" ", strip=True).replace("\n", "").replace(" ", "")
    rate = row.select_one("div.st_rate > span.icon").get_text(" ", strip=True).replace("\n", "").replace(" ", "")

    index += 1
    
    print(index, company, price, rate)
```
