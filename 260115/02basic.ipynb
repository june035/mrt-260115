# 파이썬 기초 문법
- 출력 : print() 내장함수=built in function / 내장: 파이썬 문법이 탄생될 때부터 존재
- 입력 : input("당신의 이름은?") / "당신의 이름은?" = placeholder = 입력해야할 값의 정보
    - 해당 함수를 통해서 수집한 값은 무조건 "문자열"이 된다!!

- 연산 :
    - 사칙연산: + - * / % ** 유의-> '+' : 문자 간의 연결 연산자로서도 사용 가능
    - 비교연산: < > <= >= == != (좌항 = 우항, 우항의 값을 좌항으로 삽입)

- 변수 :
    - variable 변할 수 있는 값을 넣는 공간=그릇
    - 변수명을 작명할 때 직관적, 명시적이어야 한다. 누가 봐도 한눈에 알아볼 수 있는 이름 / 숫자가 먼저 올 수 없음
    - 팁 : 스네이크 표기법(case_study), 카멜 표기법(caseStudy)

- 자료형 : 변수 안에 담을 수 있는 값의 형태. 문자열(String) " " VS 문자(Character) ' ', 정수(Inteaser,Int), 실수(Float), 논리형(Boolean)
    - 형 변화 : 자동형 변환- input(), 명시적형 변환- 수동적으로 형태를 변환. 아래 함수들
    - 자료 형태 확인 : type()
    - int() : 정수형 변환
    - float() : 실수형 변환
    - str() : 문자열 변환
    - bool() : 논리형 변환
    - list() : 리스트 자료 구조 변환
    - dict() : 딕셔너리 자료 구조 변환
    - tuple() : 튜플 자료 구조 변환
    - set() : 세트(집합) 자료 구조 변환

- 문자열 : 독립적인 함수 혹은 속성들이 많이 존재.
    - count() : 특정 문자열에서 어떤 단어가 몇 번 사용되었는지를 집계. 대/소문자 구분 필요.
    - len() : lengh. 띄어쓰기도 수량 포함. 사용 예시 - 댓글의 길이가 길수록 좋은 평점을 준다?!
    - find() : 해당 문자열에서 인자값으로 전달된 문자가 몇 번째 인덱스(0부터 시작)에 위치하는지 찾고자 할 때.
                - 문서에서 가장 처음 찾은 문자의 인덱스 반환
    - replace() : 특정 문자열을 대체하고자 할 때
    - index : 문자열 안에 각각의 문자는 고유한 숫자를 가지고 있다 -> 문자열 왼쪽부터 숫자를 세며, 0부터 시작.
    - slicing : 문자열 안에서 특정 구간을 잘라내서 가져오고 싶을 때
    - strip() : 특정 문자열 앞뒤에 여백 혹은 여백으로 인지할 수 있는 escape sequence(\n, \t) 문자열을 제거하고자 할 때
    - format() : 특정 문자열을 원하는 위치에 출력하고자 할 때 사용하는 method 함수

- 형식 지정자 : 특정 타입의 값을 출력하고자 할 때, 어떤 형태로 값을 찾아와서 출력할지 결정
    - %s - string (문자열)
    - %c - character
    - %d - int (정수)
    - %f - float (실수)

```python
text = "banana"

# print(dir(text)) 결과값에 find 가 있기 때문에 사용 가능

# idx = text.find("a")
# print(idx) --> 첫 번째 나오는 a 찾을 때

index = text.find("a")

next_index = text.find("a", index + 1)

print(index, next_index)
```

```python
# 반복문 사용

text = "banana"
target = "a"

# idx = text.find("e") 존재하지 않는 문자는 결과값이 -1
# 예외 조항 처리 : 어떤 명령문을 실행시키고자 할 때, 만약 A가 아니라면 B 해라.

idx = text.find(target)
while idx != -1 :
    print(idx)
    idx = text.find(target, idx + 1)
```

```python
text = "python"

print(text)
print(text[0])
print(text[1])
# 어떤 문자열이던, 마지막 문자의 인덱스 넘버는 전체 문자열 개수 -1 과 동일하다

print(len(text))
print(text[len(text) - 1])
print(text[-2])

# 사이트에서 리뷰를 크롤링 했을 때, 말머리로 붙은 '댓글 : ' 을 제외하고 진짜 작성된 본문부터 가져오고 싶을 때
print(text)
print(text[2])
print(text[2:4]) #찾아오고자 하는 문자열 [start:finish+1]
print(text[2:6])
print(text[2:]) # 슬라이싱에서 마지막 번째 인덱스의 생략 = 끝까지를 의미
print(text[2:len(text)]) #가장 많이 사용
print(text[2:-1])
```

```python
some_str = "  computer  "
print(some_str)
print(some_str.strip()) #strip() -> 인자값이 없는 경우가 디폴트=여백 혹은 여백과 유사한 문자열을 제거한다

some_str01 = ",,,computer..."
print(some_str01.strip(","))
print(some_str01.strip("."))
print(some_str01.strip(",."))
```

```python
name = "I have a {}, I have an {}."
print(name)

print(name.format("pen", "apple"))
print("I have a {}, I have an {}.".format("pen", "apple"))
print("I have an {1}, I have a {0}.".format("pen", "apple")) #숫자 0부터
# 매서드 함수는 문자열 형식의 객체가 반드시 있어야 한다. 함수 앞에 온점(.)이 있다면 메서드 함수.

interest = 0.087
print(format(interest, ".2f")) 
#온점이 없다면 독립함수. (문자, 숫자x) -> 소수점 위치 일치시킬 때 사용
```

```python
# 파이썬 내장함수 가운데 무조건 소수점 자리를 올리거나 내리는 독립 기능 없음
# 외부 라이브러리 모듈을 찾아와서 사용해야 함
# 그럴 때, 아래의 수학 객체 라이브러리 모듈 사용

import math

interest = 0.891
result01 = math.floor(interest) # floor : 버리다
print(result01)

result02 = math.ceil(interest) # ceil : 올리다
print(result02)
```

```python
print("I have a %s, I have an %s." % ("pen", "apple"))
print("I have a %f, I have an %s." % (3.14, "apple"))
print("I have a %f, I have an %d." % (3.14, 7))
print("I have a %f, I have an %c." % (3.14, 'A')) # 파이썬에서는 문자 " " & ' ' but 엄격한 프로그래밍 언어에서는 문자 ' ', 문자열 " "
print("I have a %f, I have an %s." % (3.14, "AB")) 
print("I have a %f, I have an %c." % (3.14, 65)) # 문자-정수는 치환된다. C++ ASCII Chart 에 따름.
```
