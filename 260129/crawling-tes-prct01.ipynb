```python
import requests
import openpyxl
from openpyxl.styles import Alignment, Font, colors
from bs4 import BeautifulSoup

wrap_alignment = Alignment(wrap_text=True)
bold_font = Font(bold=True)
header_font = Font(bold=True, size=14)
hyperlink_font = Font(color=colors.BLUE, underline="single")

excel_file = openpyxl.Workbook()
excel_sheet = excel_file.active
excel_sheet.title = "게시글"

header = ["No", "게시글 제목", "댓글"]
excel_sheet.append(header)
for cell in excel_sheet[1] :
    cell.font = header_font

excel_sheet.column_dimensions["A"].width = 10
excel_sheet.column_dimensions["B"].width = 40
excel_sheet.column_dimensions["C"].width = 80

url = "https://davelee-fun.github.io/trial/board/news.html"
res = requests.get(url)
soup = BeautifulSoup(res.text, "html.parser")

items = soup.select("div.list_item.symph_row")
index = 0
row = 2

base_url = "https://davelee-fun.github.io/trial/board/"

for item in items :
    title = item.select_one("span.subject_fixed")
    reply = item.select_one("span.rSymph05")
    link = item.select_one("a.list_subject")
    total_url = base_url+link["href"]
    
    if title != None :
        index += 1
        # print(index, title.get_text(" ", strip=True), reply.get_text(" ", strip=True), total_url)
        data1 = [index, title.get_text(" ", strip=True), "["+reply.get_text(" ", strip=True)+"]"]
        excel_sheet.append(data1)

        for cell in excel_sheet[row] :
            cell.alignment = wrap_alignment
            cell.font = bold_font

        link_cell = excel_sheet.cell(row=row, column=2)
        link_cell.hyperlink = total_url
        link_cell.font = hyperlink_font

        row += 1

        res_title = requests.get(total_url)
        soup_title = BeautifulSoup(res_title.text, "html.parser")
        reply_title = soup_title.select("div.comment_view")

        for reply in reply_title :
            # print("ㄴ", reply.get_text(" ", strip=True).replace("\n", "").replace("\t", ""))
            data2 = [" ", " ", reply.get_text(" ", strip=True).replace("\n", "").replace("\t", "")]
            excel_sheet.append(data2)
            for cell in excel_sheet[row] :
                cell.alignment = wrap_alignment

            row += 1

export_file = "mysite-prct.xlsx"
excel_file.save(export_file)
print(f"{export_file} 저장 완료")
```
