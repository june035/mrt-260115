# git 사용 방법
1) 로컬 컴퓨터(개인 PC)에 git 응용프로그램 설치
2) 터미널에서 git -v 명령어 입력해 설치 여부 확인 (새로 설치했을 때는 터미널 재실행)
- 윈도우: 검색바 > cmd 입력 > 명령 프롬프트
- 맥: cmd + space > terminal
3) 없다면, 구글 크롬에서 git 검색해서 다운로드
- 맥-home brew, 윈도우-chocolatey : 대리로 응용 프로그램을 다운로드 해 설치해주는 중간 프로그램
4) git의 소유주 확인 & 변경
- 터미널에서 git config --list 명령어 입력 email과 name 확인
- github에서 정보 확인
    - user.emali : junehg620@gmail.com
    - user.name : june035
- 터미널 마우스 우측, 관리자 모드로 재접속 > 전역(global) 세팅, 아래 명령어 입력
    - git config --global user.name "june035"
    - git config --global user.email "junehg620@gmail.com"
5) github에 회원 가입해서 그 계정 정보 값으로 입력
- git : 나의 로컬 컴퓨터에서 데이터 버전을 관리해주는 툴
- github : 나의 로컬 컴퓨터에서 관리하고 있는 버전을 타인과 공유해서 사용하고자 할 때, 내가 사용하고 있는 디바이스가 2대 이상인 경우, A와 B 모두에서 데이터를 관리하고자 할 때
- repository : github > project에 따른 개별 폴더
6) git commit & push (데이터 업로드)
- 루트 폴더 경로 확인 (C드라이브 > USER > dataclass_weekday)
- 루트 폴더 지정: 터미널에서 cd "현재 나의 루트 폴더" 명령어 입력 / change dircetory
    - cd C:\Users\user\dataclass_weekday  
- git 초기화 : 터미널에서 git init 명령어 입력 > 루트 폴더로 가서 '숨김 파일 보기' 설정, .git 파일 생성 확인
- staging area : 데이터 전송 : 터미널에서 git add . 명령어 입력
- 데이터가 정상적으로 tracking 됐는지 확인 : 터미널에서 git status 명령어 입력
- 데이터 labeling 하기 : 터미널에서 git commit -m "260115"
- 원격의 공간(웹)인 github 레포지토리에서 경로 확인
    - git remote add origin https://github.com/june035/mrt-260115.git
- 로컬(git)과 원격(github) 이 연결됐는지 확인 : 터미널에서 git remote -v
    - fetch : 데이터를 받아올 곳
    - push : 데이터를 보낼 곳
- 로컬 -> 원격으로 데이터 push : 터미널에서 git push origin master
    - github 새로고침해서 레포지토리에 폴더 생성됐는지 확인
- permission 이 안됐을 경우. 찾기 - '자격' 검색 - 자격 증명 관리 - windows 자격증명 - git 자격 확인 - 기존꺼 제거 - 터미널에서 다시 git push origin master
7) git clone & pull (데이터 다운로드)
- 로컬 컴퓨터에
- git 설치
- 새로운 루트 폴더(예: home-pc) 만들고 경로 복사
- 터미널로 가서 cd C:\Users\user\home-pc
- github에서 레포지토리 경로 확인, 터미널에서 git clone https://github.com/june035/mrt-260115
- 루트 폴더에서 mkt-260115 생성됐는지 확인
- 집에서 학원 파일을 오픈해서 추가로 내용 업데이트해서 다시 레포지토리에 푸쉬하려면, staging area(명령어 git add .)부터 진행
8) git fork
