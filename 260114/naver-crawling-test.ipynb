```python
# 네이버 웹포털 사이트 내 뉴스 카테고리에서 특정 키워드 크롤링 해오기
# 크롤링 (crawling) : 외부에서 데이터를 수집해오는 행위
# ctrl + / : 주석처리 가능, ctrl + z : 직전 단계로 돌아가기

import requests 
from bs4 import BeautifulSoup

query = "마케팅" # 변수 선언 후 값을 할당
url = f"https://search.naver.com/search.naver?where=news&ie=utf8&sm=nws_hty&query={query}" # f-string 문법

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/143.0.0.0 Safari/537.36"
}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

titles = soup.select("a.fender-ui_228e3bd1.MNH2zYqVX0VCatL92VdS")

for i, title in enumerate(titles[0:10], 1) :
    print(f"{i}. {title.text}")


# bs4.BeautifulSoup()
# print(response) #print 함수 => 어떤 기능을 구현하고자 할 때 사용하는 문법
# 방법은 여러가지. 숫자로 리스트업할 때 1) i+1 2) [0:10], 1
```
