```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time 

# 로컬 컴퓨터의 브라우저 드라이브와 동일한 버전으로 설치 & 객체 생성
service = Service(ChromeDriverManager().install())

# 가상 브라우저가 실제처럼 보일 수 있도록 설정하는 옵션
options = Options()
options.add_argument("--headless") #가상 브라우저가 띄워졌음에도 우리 눈에는 나타나지 않음
options.add_argument("--disable-gpu") # headless 했다면 gpu 작동 불필요
options.add_argument("--disable-dev-shm-usage") # 가상 메모리 사용 제한
options.add_argument("--windonw-size=1920,1080") # 가상 브라우저 사이즈 정보 설정
options.add_argument("--start-maximized") # 반응형 사이트에서도 전체 사이즈(데스크탑)에 맞춰라 
options.add_argument("--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36")  
options.add_argument("--lang=ko_KR")
options.add_argument("--no-sandbox") # sandbox = 규제. 크롤링 자체를 정책적으로 차단할 때 우회해서 들어갈 수 있음

# 브라우저 드라이버 정보값을 가지고 있는 객체를 매개변수로 받아서 크롤링
driver = webdriver.Chrome(service=service)
driver.get("https://davelee-fun.github.io/")
# time.sleep(4)

elements = driver.find_elements(By.CSS_SELECTOR, "div.wrapthumbnail img")

for element in elements :
    print(element.get_attribute("src"))


driver.quit()
```

### selenium 문법

1. find_elements() : 복수의 값을 찾아온다. 동일한 요소를 전체 다 찾아서 가져온다 -> 리스트
2. find_elemenet() : 단수. 위에서부터 문서를 읽다가 최초로 발견한 값을 찾아오고 종료
   - requests에서 find(단일 태그를 중심), find_all(복수 태그), select(복수의 선택자: . #), select_one(단일 선택자)
3. By : .TAG_NAME 태그 중심으로 찾아온다 .ID 아이디 중심->문서 상에서 1번 밖에 못 쓴다(따라서 find_element 적합) .CSS_SELECTOR 선택자 중심(클래스, 태그, ID) .NAME 'name' 속성의 값을 찾아옴. 'name'은 식별 역할.
4. get_attribute() : 특정 속성의 속성값을 찾아오고자 할 때
5. .clear() : 사용자가 값을 입력하도록 유도하는 특정 영역(Form 요소 기준> input, textarea 외) 내 제시된 값을 삭제
6. .send_keys() : 어떤 값 혹은 키보드의 키값을 전달하고자 할 때
7. Keys : 키보드의 키를 입력하고자 할 때
8. sleep(n) : n초의 시간동안 지연시키고자 할 때
9. screenshot(filename) : filename으로 스크린 캡쳐 가능
10. set_window_size(가로, 세로) : 윈도우 창 사이즈 조정

```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys 
from selenium.webdriver.chrome.options import Options
import time 

service = Service(ChromeDriverManager().install())

options = Options()
# options.add_argument("--headless") 가상 브라우저가 띄워졌음에도 우리 눈에는 나타나지 않음
# options.add_argument("--disable-gpu") headless 했다면 gpu 작동 불필요
options.add_argument("--disable-dev-shm-usage") 
options.add_argument("--windonw-size=1920,1080") 
options.add_argument("--start-maximized") 
options.add_argument("--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36")  
options.add_argument("--lang=ko_KR")
options.add_argument("--no-sandbox")

driver = webdriver.Chrome(service=service)
driver.get("https://davelee-fun.github.io/")

driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")
# 스크롤을 할 수 있는 건 javascript뿐. execute_script: 자바스크립트 언어를 빌려 쓰는 것
# scrollTo(0, document.body.scrollHeight) : 현재 문서에서 내려갈 수 있는 만큼 최대한 스크롤을 내려라

element = driver.find_element(By.NAME, "EMAIL")
element.clear() #placeholder 내부값(속성값)을 젲거
element.send_keys("abcj@gmail.com") #NAME 속성값에 " " 넣어라

time.sleep(4)

element.send_keys(Keys.RETURN) #엔터키(RETURN)를 눌러라 = 이메일 넣고 구독하기 눌러라

time.sleep(4)

driver.quit() 
```

```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys 
from selenium.webdriver.chrome.options import Options
import time 

service = Service(ChromeDriverManager().install())

options = Options()
# options.add_argument("--headless") 가상 브라우저가 띄워졌음에도 우리 눈에는 나타나지 않음
# options.add_argument("--disable-gpu") headless 했다면 gpu 작동 불필요
options.add_argument("--disable-dev-shm-usage") 
options.add_argument("--windonw-size=1920,1080") 
options.add_argument("--start-maximized") 
options.add_argument("--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36")  
options.add_argument("--lang=ko_KR")
options.add_argument("--no-sandbox")

driver = webdriver.Chrome(service=service)
driver.get("https://davelee-fun.github.io/")

driver.set_window_size(144, 1000)
element = driver.find_element(By.TAG_NAME, "body")

element.screenshot("crawlingsite.png")

time.sleep(4)

driver.quit() 
```
