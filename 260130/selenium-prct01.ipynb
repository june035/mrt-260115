### selenum 문법
1. find_elements() : 복수의 값을 찾아온다. -> 리스트
2. find_element() : 단수의 값을 찾아온다
3. By.TAG_NAME, "" : 태그 중심으로 찾아온다
4. By.ID, "" : ID 중심으로 찾아온다 -> 문서 상에서 1번 밖에 못 쓰므로 find_element에 적합
5. By.CSS_SELECTOR, "" : 선택자 중심으로 찾아온다(클래스, 태그, ID)
6. By.NAME, "" : 'name' 속성의 값을 찾아온다. 'name'은 식별 역할
7. get_attribute("") : 특정 속성의 속성값을 찾아오고자 할 때
8. .clear() : 사용자가 값을 입력하도록 유도하는 특정 영역 내 제시된 값을 삭제
9. .send_keys() : 어떤 값 혹은 키보드의 키값을 전달하고자 할 때
10. Keys : 키보드의 키를 입력하고자 할 때
11. time.sleep(n) : n초의 시간동안 지연시키고자 할 때
12. screenshot("filename") : filename으로 스크린 캡쳐 가능
13. set_window_size(가로, 세로) : 윈도우 창 사이즈 조정

```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time

service = Service(ChromeDriverManager().install())

options = Options()
options.add_argument("--headless")
options.add_argument("--disable-gpu")
options.add_argument("--disable-dev-shm-usage")
options.add_argument("--window-size=1920,1080")
options.add_argument("--start-maximized")
options.add_argument("--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36")
options.add_argument("--lang=ko_KR")
options.add_argument("--no-sandbox")

driver = webdriver.Chrome(service=service)
driver.get("https://davelee-fun.github.io/")
time.sleep(4)

elements = driver.find_elements(By.CSS_SELECTOR, "div.wrapthumbnail img")

for e in elements :
    print(e.get_attribute("src"))

driver.quit()
```

```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

service = Service(ChromeDriverManager().install())

options = Options()
options.add_argument("--headless")
options.add_argument("--disable-gpu")
options.add_argument("--disable-dev-shm-usage")
options.add_argument("--window-size=1920, 1080")
options.add_argument("--start-maximized")
options.add_argument("--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36")  
options.add_argument("--lang=ko_KR")
options.add_argument("--no-sandbox")

driver = webdriver.Chrome(service=service)
driver.get("https://davelee-fun.github.io/")

driver.set_window_size(144, 1000)
element = driver.find_element(By.TAG_NAME, "body")

element.screenshot("crawlingsite-prct.png")

time.sleep(4)

driver.quit()
```
