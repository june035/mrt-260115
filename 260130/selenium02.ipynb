```python
# 미션
"""
1) 구글 scholar 사이트 접속
2) 검색바에 SEO Marketing 키워드 자동 입력
3) 검색 자동 실행
4) 논문 제목 5개 크롤링한 후 출력
"""

from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys 
from selenium.webdriver.chrome.options import Options
import time 

service = Service(ChromeDriverManager().install())

options = Options()
# options.add_argument("--headless") 
# options.add_argument("--disable-gpu")
options.add_argument("--disable-dev-shm-usage") 
options.add_argument("--windonw-size=1920,1080") 
options.add_argument("--start-maximized") 
options.add_argument("--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36")  
options.add_argument("--lang=ko_KR")
options.add_argument("--no-sandbox")

driver = webdriver.Chrome(service=service)
driver.get("https://scholar.google.com/?hl=ko")

element = driver.find_element(By.NAME, "q")
element.clear() 
element.send_keys("SEO Marketing") 

element.send_keys(Keys.RETURN)

time.sleep(2)

elements = driver.find_elements(By.TAG_NAME, "h3")
for title in elements :
    print(title.text)

time.sleep(8)

driver.quit()
```
