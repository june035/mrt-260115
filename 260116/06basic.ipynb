    - strip() : 특정 문자열 앞뒤에 여백 혹은 여백으로 인지할 수 있는 escape sequence(\n, \t) 문자열을 제거하고자 할 때
    - format() : 특정 문자열을 원하는 위치에 출력하고자 할 때 사용하는 method 함수

- 형식 지정자 : 특정 타입의 값을 출력하고자 할 때, 어떤 형태로 값을 찾아와서 출력할지 결정
    - %s - string (문자열)
    - %c - character
    - %d - int (정수)
    - %f - float (실수)

```python
# 0부터 시작해서 10번 반복
for index in range(10) :
    print(index) 
```

```python
# 1부터 시작해서 10번 반복. 문자열의 슬라이싱과 유사
for index in range(1, 11) :
    print(index)
```

복합대입연산자
- a = a + b -> a += b
- a = a - b -> a -= b
- a = a * b -> a *= b
- a = a / b -> a /= b
- a = a % b -> a %= b
- a = a ** b -> a **= b

```python
# 누산기 (누적 합산)
sum = 0 # 어떤 값이든 영향을 미치지 않는 값 = 0 으로 설정

for index in range(1, 11) :
    # sum = sum + index
    sum += index #복합대입연산자. 위의 구문과 동일

print(sum)
```

```python
name = ""

while name != "David" :
    name = input("당신의 이름은?")
    if name != "David" : 
        print(f"HI {name}, You know David?")
    else :
         print(f"HI {name}")
```

```python
# 사용자에게 1개의 숫자를 받아서 구구단을 출력하세요.

num = int(input("몇 단의 구구단을 출력할까요?"))
for digit in range(1, 10) :
    print(f"{num} x {digit} = {num * digit}")
```

```python
# Q. 1~30까지의 숫자 중 3의 배수만 출력해주세요.

num = 0

while num < 30 :
    num += 1
    if num % 3 == 0 :
        print(num)
```
