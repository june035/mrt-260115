# 조건문 : 값을 출력하기 위해서 A | B 경우에 따라서 다른 값을 출력해야할 때

1) if 조건문 : A라면 ~ 해라 / B라면 ~ 해라 (단일 조건)
2) if else문 : A라면 ~ 해라, 만약 A가 아니라면 - 해라 (양자택일)
3) if elif else문 : A라면 ~ , B라면 ~ , C라면 ~ (다중택일)
> else 문 : 무조건 if, elif 보다 먼저 나올 수 없다
> else 문 : 필수 조건이 아닌 비필수 조건. if elif else -> if elif
> 조건문에 자주 사용하는 문법 : 조건식 + 실행문(들여쓰기 필수)
    ㄴ 조건문의 조건식에 비교연산자, 논리연산자가 주로 사용됨
- 중첩 if 조건문

```python
# Q. 사용자에게 3개의 숫자를 입력받은 후 그 중 작은 숫자를 출력하세요.

num01 = int(input("첫번째 숫자"))
num02 = int(input("두번째 숫자"))
num03 = int(input("세번째 숫자"))

if num01 <= num02 and num01 <= num03 :
    print(num01)
elif num02 <= num01 and num02 <= num03 :
    print(num02)
elif num03 <= num01 and num02 <= num02 :
    print(num03)
```

```python
numbers = [] #값이 들어오는 공간, 리스트인 이유-추가해야할 수도 있으니까

for i in range(3) :
    number = int(input(f"{i+1}번째 숫자 : "))
    numbers.append(number)

# 구조분해할당 문법 : 리스트와 같이 여러 개 변수값을 가지고 있는 자료 구조의 값들을 하나씩 추출해서 독립 변수로 담을 수 있는 방법

num01, num02, num03 = numbers

if num01 <= num02 and num01 <= num03 :
    print(num01)
elif num02 <= num01 and num02 <= num03 :
    print(num02)
elif num03 <= num01 and num02 <= num02 :
    print(num03)
```

```python
numbers = []

for i in range(3) : #반복'문'(식+식), ':' 문장으로 쓰겠다
    number = int(input(f"{i+1}번째 숫자 : "))
    numbers.append(number)

print(min(numbers))
```

```python
# 리스트 컴프리핸션 = list comprehension

numbers = [int(input(f"{i + 1}번째 숫자 : ")) for i in range(3)]
# 리스트의 형태를 가졌지만 실제 리스트의 속성값을 모두 사용할 수 없음
# sort, reverse -> 원래 값을 편집하는 속성 때문에 인덱스 구조가 사라진다

print(numbers)
print(min(numbers))
print(sorted(numbers)[0])
```
